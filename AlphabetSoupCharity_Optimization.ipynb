```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_copy1 = application_df.copy()
```

```python
application_copy2 = application_df.copy()
```

```python
application_copy3 = application_df.copy()
```

Optimization #1, binning names

```python
# Drop the non-beneficial ID column 'EIN'.
application_copy1 = application_copy1.drop(columns=['EIN'])
```

```python
# Determine the number of unique values in each column.
application_copy1.nunique()
```

```python
name_counts = application_copy1['NAME'].value_counts()
name_counts
```

```python
name_counts[name_counts >10]
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
name_types_to_replace = name_counts[name_counts <= 10].index.tolist()

# Replace in dataframe
for app in name_types_to_replace:
    application_copy1['NAME'] = application_copy1['NAME'].replace(app,"Other_Name")

# Check to make sure replacement was successful
application_copy1['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
value_counts = application_copy1['APPLICATION_TYPE'].value_counts()
value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = value_counts[value_counts < 500].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_copy1['APPLICATION_TYPE'] = application_copy1['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_copy1['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
class_counts = application_copy1['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = class_counts[class_counts < 1000].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_copy1['CLASSIFICATION'] = application_copy1['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_copy1['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_dummy = pd.get_dummies(application_copy1)
```

```python
# Split our preprocessed data into our features and target arrays
X = application_dummy.drop(columns = 'IS_SUCCESSFUL').values
y = application_dummy.IS_SUCCESSFUL.values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input = len(X_train[0])
hidden1 = 70
hidden2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden1, input_dim=input, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
